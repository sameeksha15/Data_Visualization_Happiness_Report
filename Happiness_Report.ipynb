## Stastics of Counries with Top 10 Happiness Ranking

```python
top10 = df[df['Happiness Rank'] <= 10]
```

```python
top10
```

```python
top10['Region'].unique()
```

```python
region_map = {'Western Europe': 1, 
              'North America': 2, 
              'Australia and New Zealand': 3
             }
```

```python
top10_pie = top10.copy()
```

```python
top10_pie['Region'] = top10_pie['Region'].replace(region_map)
```

```python
top10_pie
```

```python
regions = ['Western Europe', 'Australia and New Zealand',  'North America']
```

```python
top10_pie['Region'].value_counts()
```

```python
top10_pie['Region'].value_counts().sum()
```

```python
we_per = (top10_pie[top10_pie['Region'] == 1].value_counts().sum() / top10_pie['Region'].value_counts().sum()) * 100
aus_nz_per = (top10_pie[top10_pie['Region'] == 3].value_counts().sum() / top10_pie['Region'].value_counts().sum()) * 100
na_per = (top10_pie[top10_pie['Region'] == 2].value_counts().sum() / top10_pie['Region'].value_counts().sum()) * 100
```

```python
plt.figure(figsize=(15, 6))

plt.pie(top10_pie['Region'].value_counts(), explode=[0.1, 0.0, 0.2], shadow =True, startangle=90)
plt.title('Distribution of regions  of countries with Top 10 Happiness Ranking')
plt.legend(regions, loc ="center left", bbox_to_anchor =(1, 0, 0.5, 1))
plt.savefig('img/pie1.png')
```

## Stastics of Countries with last (bottom) 10 Happiness Ranking

```python
## Last 10 happiest countries
last10 = df.tail(10)
last10
```

```python
last10['Region'].unique()
```

```python
least_happiest_map = {
    'Sub-Saharan Africa': 1,
    'Southern Asia': 2,
    'Middle East and Northern Africa': 3
}
```

```python
least10_pie = last10.copy()
```

```python
least10_pie['Region'] = least10_pie['Region'].replace(least_happiest_map)
least10_pie
```

```python
least10_pie['Region'].value_counts()
```

```python
least_happy_regions = ['Sub-Saharan Africa', 'Southern Asia', 'Middle East and Northern Africa']
```

```python
ssa_per = (least10_pie[least10_pie['Region'] == 1].value_counts().sum() / least10_pie['Region'].value_counts().sum()) * 100
sa_per = (least10_pie[least10_pie['Region'] == 2].value_counts().sum() / least10_pie['Region'].value_counts().sum()) * 100
me_na_per = (least10_pie[least10_pie['Region'] == 3].value_counts().sum() / least10_pie['Region'].value_counts().sum()) * 100
```

```python
plt.figure(figsize=(15, 6))

plt.pie(least10_pie['Region'].value_counts(), explode=[0.1, 0.0, 0.2], shadow =True, startangle=90)
plt.legend(least_happy_regions, loc ="center left", bbox_to_anchor =(1, 0, 0.5, 1))
plt.title('Distribution of regions  of countries with Bottom 10 Happiness Ranking')
plt.savefig('img/pie2.png')
```

## Comparision between the Economy (GDP per Capita) of Top 10 and Bottom 10 Countries

```python
temp1 = top10.copy()
```

```python
temp1 = temp1.sort_values(by='Economy (GDP per Capita)', ascending=True)
temp1
```

```python
temp2 = last10.copy()
```

```python
temp2 = temp2.sort_values(by='Economy (GDP per Capita)', ascending=True)
temp2
```

```python
plt.figure(figsize=(12, 6))


plt.subplot(121)
sns.barplot(x=temp1['Country'], y=temp1['Economy (GDP per Capita)'].sort_values())
plt.title('GDP per Capita of countries with Top 10 Happiness Ranking')
plt.xticks(rotation=90)


plt.subplot(122)
sns.barplot(x=temp2['Country'], y=temp2['Economy (GDP per Capita)'])
plt.xticks(rotation=90)
plt.title('GDP per Capita of countries with Bottom 10 Happiness Ranking')
plt.savefig('img/bar2.png')
```
